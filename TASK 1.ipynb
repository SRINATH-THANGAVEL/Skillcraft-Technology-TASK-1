```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\welcome\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_127006.csv",skiprows=4)

print(df.head())
print(df.columns)

india_data = df[df['Country Name'] == 'India']

years = india_data.columns[4:]
population_values = india_data[years].values.flatten()

plt.figure(figsize=(12,6))
plt.bar(years, population_values, color='skyblue')
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("India's Population over Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
